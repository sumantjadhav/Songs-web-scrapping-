### In this project I have scrapped 'IMDB' website for songs

### Library used is BeautifulSoup

```python
#import all the necessary Libraries
from bs4 import BeautifulSoup
import requests,openpyxl
import os
```

### Creating a Excel file to store the data

```python
excel= openpyxl.Workbook()
sheet=excel.active
sheet.title="Songs ratings"
sheet.append(['Rank','name','time','year','rating'])
```

### Using 'try and except' in order to eliminate wrong url errors

```python
try:
    source= requests.get('https://www.imdb.com/search/title/?title_type=video&genres=music&sort=user_rating,desc&ref_=adv_prv')
    source.raise_for_status()
    soup=BeautifulSoup(source.text,'html.parser')
    print(soup)
    
except  Exception as e:
    print(e)
```

```python
songs=soup.findAll('div', attrs={"class","lister-item mode-advanced"})
print(songs)

```

### Using "for" loop to iterate through the page

```python
for song in songs:
    rank=song.find('span',class_='lister-item-index unbold text-primary').text.strip(".")
    
    name=song.find('h3', class_='lister-item-header').a.text
    
    time=song.find('p', class_='text-muted').span.text
    
    year=song.find('span',class_='lister-item-year text-muted unbold').text.strip('()')
    
    raiting=song.find('strong').text
    
    sheet.append([rank,name,time,year,raiting])
    
    
   
    
  
    
```
